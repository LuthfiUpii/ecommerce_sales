Soal 01

Pada bulan apakah perusahaan memiliki pendapatan tertinggi dan terendah?
Berapa jumlah pendapatan pada bulan tersebut? Dari hasil analisa, coba
visualisasikan dan berikan pendapat Anda dari visualisasi tersebut.

**Import** **library**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Read file csv**

```python
df = pd.read_csv("sample_data/ecommerce_sales.csv")
df
```

**Melihat total semua data Nan**

```python
df.isna().sum()
```

**drop semua data NaN, dan di simpan secara permanen**

```python
df.dropna(inplace=True)
```

**Drop semua data yang sama**

```python
df.drop_duplicates(inplace=True)
```

**Mengecek data apakah masih ada data yang Nan**

```python
df.isna().sum()
```

**mengembalikan satu nilai untuk setiap kolom**

```python
data = df[(df != df.columns).all(axis=1)]
data
```

**Mengubah type object menjadi float agar bisa di kalikan antara quantity dan price**

```python
data["Quantity Ordered"] = data["Quantity Ordered"].astype(float)
data["Price Each"]=data["Price Each"].astype(float)
data
```

**menambah column month baru dengan type int dari order date untuk menggambil dari bulan saja dengan menggunakan idex 0:2**

```python
data['Month'] = data['Order Date'].str[0:2]
data['Month'] = data['Month'].astype('int')
data
```

**menambah colum baru untuk hasil dari perkalian dari quantity dan price**

```python
data["Total Paid"] = data['Quantity Ordered'] * data['Price Each']
data
```

**menjumlahkan total pendapatan dari setiap bulannya, dan terlihat pendapatan tertinggi adalah pada bulan desember**

```python
Pendapatan = data['Total Paid'].groupby(data['Month']).sum()
Pendapatan
```

```python
Pendapatan.max()
```

# **Dari hasil analisa penjualan perbulan yang dibuatkan dalam plot bar, terlihat pendapatan tertinggi ada di bulan Desember dan terendah ada pada bulan Januari.**

```python
Pendapatan.plot(kind='bar',x='Month',title='Pendapatan Tertinggi Perbulan')
plt.ylabel('Total Paid')
plt.show()
```
